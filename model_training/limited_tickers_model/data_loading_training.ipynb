```python
import polars as pl

df = pl.read_csv(dir + "/articles_with_11day_ohlc_local.csv")
df
```

```python
df = df.select(["Article_title", "Stock_symbol", "Article_Date", "Close_-3", "Close_-2", "Close_-1", "Close_0", "Close_1", "Close_2", "Close_3"])
```

```python
df = df.with_columns([
    (
        (pl.col("Close_1") / pl.col("Close_-1"))
        .log()
    ).alias("ret_m1_to_p2_log"),
])
```

```python
threshold = 0.0075

df = df.filter(pl.col("ret_m1_to_p2_log").abs() > threshold)

df = df.with_columns(
    pl.when(pl.col("ret_m1_to_p2_log") > 0)
      .then(1)
      .otherwise(0)
      .alias("stock_movement")
)
```

```python
df = df.rename({"Article_title": "headline"})

df = df.with_columns(
    (pl.col("Stock_symbol") + ": " + pl.col("headline")).alias("headline")
)

df
```

```python
grouped_df = df.group_by("Stock_symbol").agg(
        pl.len().alias("n_events")
  )
good_syms = grouped_df.filter(pl.col("n_events") >= 2000)["Stock_symbol"]
```

```python
half_good_syms = good_syms.sample(n=10)
```

```python
df_filtered = df.filter(pl.col("Stock_symbol").is_in(half_good_syms))
df_filtered
```

```python
print(len(df_filtered))
```

```python
df_filtered
```

```python
train_df = df_filtered.to_pandas()

train_df["labels"] = train_df["stock_movement"]
train_df.dropna(subset=["labels"], inplace=True)  # in case any weird rows
train_df.drop("stock_movement", axis=1, inplace=True)
```

```python
#LORA
from datasets import Dataset
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    train_df["headline"],
    train_df["labels"],
    test_size=0.2,
    random_state=42,
)

train_df_hf = (
    {"headline": list(X_train), "labels": list(y_train)}
)
val_df_hf = (
    {"headline": list(X_val), "labels": list(y_val)}
)

train_ds = Dataset.from_dict(train_df_hf)
val_ds   = Dataset.from_dict(val_df_hf)

from transformers import AutoTokenizer

base_model_name = "roberta-base"
tokenizer = AutoTokenizer.from_pretrained(base_model_name)

def tokenize_fn(batch):
    return tokenizer(
        batch["headline"],
        truncation=True,
        max_length=128,
    )

train_tok = train_ds.map(tokenize_fn, batched=True)
val_tok   = val_ds.map(tokenize_fn, batched=True)

train_tok = train_tok.remove_columns(["headline"])
val_tok   = val_tok.remove_columns(["headline"])
from transformers import (
    AutoModelForSequenceClassification,
    TrainingArguments,
    Trainer,
    DataCollatorWithPadding,
)

model = AutoModelForSequenceClassification.from_pretrained(
    base_model_name,
    num_labels=2,
)

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

training_args = TrainingArguments(
    output_dir= dir+ "/roberta_binary",
    num_train_epochs=10,
    per_device_train_batch_size=16,
    gradient_accumulation_steps=4,
    learning_rate=2e-5,
    logging_steps=200,
    eval_strategy="epoch",
    save_strategy="no",
    fp16=True,
    report_to="none",
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_tok,
    eval_dataset=val_tok,
    data_collator=data_collator,
)

# trainer.train()
trainer.train(resume_from_checkpoint=True)
metrics = trainer.evaluate()
print(metrics)


pred = trainer.predict(val_tok)
logits = pred.predictions
labels = pred.label_ids

# For binary classification, class 1 is "up", class 0 is "down"
preds = np.argmax(logits, axis=-1)

accuracy = (preds == labels).mean()
print("RoBERTa accuracy:", accuracy)

```

```python
def print_trainable_parameters(model):

    # Count parameters in each component
    encoder_total = sum(p.numel() for p in model.roberta.parameters())
    encoder_trainable = sum(p.numel() for p in model.roberta.parameters() if p.requires_grad)

    head_total = sum(p.numel() for p in model.classifier.parameters())
    head_trainable = sum(p.numel() for p in model.classifier.parameters() if p.requires_grad)

    total = sum(p.numel() for p in model.parameters())
    trainable = sum(p.numel() for p in model.parameters() if p.requires_grad)

    print(f"{'Component':<20} {'Trainable':<15} {'Total':<15} {'Status'}")
    print("-" * 65)
    print(f"{'Encoder (roberta)':<20} {encoder_trainable:<15,} {encoder_total:<15,} {'🔓 Trainable' if encoder_trainable > 0 else '🔒 Frozen'}")
    print(f"{'Head (classifier)':<20} {head_trainable:<15,} {head_total:<15,} {'🔓 Trainable' if head_trainable > 0 else '🔒 Frozen'}")
    print("-" * 65)
    print(f"{'TOTAL':<20} {trainable:<15,} {total:<15,} {f'{trainable/total*100:.1f}% trainable'}")

base_model_name = "roberta-base"
model_to_classify = AutoModelForSequenceClassification.from_pretrained(
    base_model_name,
    num_labels=3,
)

print_trainable_parameters(model_to_classify)

lora_config = LoraConfig(
  task_type=TaskType.SEQ_CLS,
  r=16,
  lora_alpha=32,
  lora_dropout=0.1,
  # target_modules=["query", "value"],   # RoBERTa attention
  modules_to_save=["RobertaClassificationHead"]
)

lora_model = get_peft_model(model_to_classify, lora_config)

print_trainable_parameters(lora_model)


```
